## helper function

```python
def get_ans(data):
    answer_start_idx = data['answers']['answer_start'][0]
    answer_end_idx = answer_start_idx + len(data['answers']['text'][0])
    print("Actual Ans >> ", data['answers']['text'])
    print("Extracted Ans >> ", data['context'][answer_start_idx : answer_end_idx])
    print("Staring index & End index >>", {answer_start_idx : answer_end_idx})

```

```python
get_ans(train_data[0])
```
